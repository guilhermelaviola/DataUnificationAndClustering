# **Creating Interactive Dashboards**
Dashboards serve as essential visual tools that consolidate data from various sources to support decision-making across operational, tactical, and strategic levels. By utilizing powerful platforms like Power BI, Tableau, or Databricks, organizations can transform raw data into interactive visualizations—such as bar, line, and geographic charts—that prioritize clarity and relevance. The process of creating an effective dashboard requires careful data preparation, a user-centric design that employs intuitive layouts, and robust interactivity through filters and menus to allow for deep exploration. Ultimately, a well-constructed dashboard not only provides real-time insights but also fosters collaboration through easy publishing and sharing, enabling diverse users to derive tailored insights for informed action.

## **Example: Data Preparation (The 'Lakehouse' Foundation)**
In Databricks, dashboards are most effective when built on top of a Silver or Gold layer table.

```python
# Importing all the necessary libraries and resources:
from pyspark.sql.functions import col, expr
from pyspark.sql import SparkSession
```

```python
# Creating a mock sales dataset as per the lesson instructions:
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SalesDashboard").getOrCreate()

data = [
    ('2024-01-01', 'North', 'Electronics', 1200, 5),
    ('2024-01-02', 'South', 'Home Office', 450, 2),
    ('2024-01-03', 'North', 'Electronics', 800, 3),
    ('2024-01-04', 'West', 'Furniture', 2100, 1),
    ('2024-01-05', 'East', 'Electronics', 300, 10),
]

columns = ['date', 'region', 'category', 'revenue', 'units_sold']
df = spark.createDataFrame(data, columns)

# Save as a table for the Dashboard to query
df.write.mode('overwrite').saveAsTable('sales_dashboard_demo')
```

## **Example: Creating Visualizations (The "Notebook" approach)**
Instead of just a static chart, we use SQL to satisfy the "Updating" and "Relevance" requirements

```python
display(spark.sql("""
SELECT
  region,
  SUM(revenue) as total_revenue,
  SUM(units_sold) as total_units
FROM sales_dashboard_demo
WHERE region = :Region_Parameter  -- This creates the 'Interactive Widget'
GROUP BY region
""", args={"Region_Parameter": "North"}))
```

# **Creating Charts & Dashboards with Databricks**
This comprehensive module guides students through the journey of data storytelling, starting with the **fundamental principles** of chart creation—such as selecting the appropriate format (bar, line, or pie) and applying visual design basics like color theory and labeling. The curriculum then advances into technical implementation within **Databricks**, where learners transition from static visuals to **interactive charts** using libraries like Plotly and progress further to compiling these visualizations into cohesive, real-time **dashboards**. Finally, the module emphasizes **professional ethics and accessibility**, teaching students how to maintain data integrity, avoid visual distortion, and ensure clarity to effectively communicate insights to diverse audiences.

```python
# Importing all the necessary libraries and resources:
from pyspark.sql import SparkSession
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Initializing SparkSession:
spark = SparkSession.builder.appName('ChartFundamentals').getOrCreate()

# Loading sample data:
df = spark.read.csv('dbfs:/data_path.csv', header=True, inferSchema=True)
df_pd = df.toPandas()

# Creating bar chart:
df_pd.plot(kind='bar', x='category', y='value')
plt.show()

# Creating line chart:
df_pd.plot(kind='line', x='time', y='value')
plt.show()

# Creating pie chart:
df_pd.groupby('category').sum().plot(kind='pie', y='value', autopct='%1.1f%%')
plt.show()
```

```python
# Initializing SparkSession:
spark = SparkSession.builder.appName('InteractiveCharts').getOrCreate()

# Loading sample data:
df = spark.read.csv('dbfs:/data_path.csv', header=True, inferSchema=True)
df_pd = df.toPandas()

# Creating interactive charts with Plotly:
fig = px.scatter(df_pd, x='category', y='value', color='category')
fig.show()
```

```python
# Initializing SparkSession:
spark = SparkSession.builder.appName('DatabricksDashboards').getOrCreate()

# Loading sample data:
df = spark.read.csv('dbfs:/data_path.csv', header=True, inferSchema=True)
df_pd = df.toPandas()

# Creating bar chart:
fig1 = px.bar(df_pd, x='category', y='value', title='Bar Chart')

# Creating line chart:
fig2 = px.line(df_pd, x='time', y='value', title='Line Chart')

# Creating heatmap:
fig3 = px.imshow(df_pd.corr(), title='Heatmap')

# Organizing visualizations in a dashboard layout:
fig1.show()
fig2.show()
fig3.show()
```

```python
# Initializing SparkSession:
spark = SparkSession.builder.appName('BestPracticesDashboards').getOrCreate()

# Loading sample data:
df = spark.read.csv('dbfs:/data_path.csv', header=True, inferSchema=True)
df_pd = df.toPandas()

# Configuring a consistent layout:
fig1 = px.bar(df_pd, x='category', y='value', title='Bar Chart')
fig2 = px.line(df_pd, x='time', y='value', title='Line Chart')
fig3 = px.imshow(df_pd.corr(), title='Heatmap')

# Picking an harmonious color palette:
px.colors.qualitative.Plotly

# Implementing interactivity with Plotly:
fig = px.scatter(df_pd, x='category', y='value', color='category', title='Interactive Dashboard')
fig.update_layout(
    xaxis_title='Category',
    yaxis_title='Value'
)
fig.show()
```
